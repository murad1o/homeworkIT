# Ход работы
<a name = 'Ход_работы'>

<p  align="justify">  1. Для решения данной задачи был установлен GIT. (см Рис. 1) </p>

![Установка Git Bash](https://cdn.discordapp.com/attachments/1040640165997191189/1174047167825059840/image.png?ex=65662ba1&is=6553b6a1&hm=d6ac9ba644db383b083426b51fa87d2aaf4777ec324178953710f59f7ed6b41a& "Рисунок 1.")




<p  align="justify">  2. Для того, чтобы выполнить данную задачу был инициализирован новый Git репозиторий, в выбранной директории на компьютере, используя команду git init. (см. Рис 2)
 </p>

![Инициализация Git репозитория](https://cdn.discordapp.com/attachments/1040640165997191189/1172954400453570630/image.png?ex=656231e9&is=654fbce9&hm=5bd29c11ff66c73240eb7cb492a792a93cfbe2d1643c1644ba27c34a0170a13f& "Рисунок 2.")




<p  align="justify">  3. Для того, чтобы выполнить данную задачу, был создан файл отчета по выполненной практической работе и помещен в репозиторий с помощью команды: "git add <путь до файла>"  (см. Рис 3)
</p>

![Добавление файла](https://cdn.discordapp.com/attachments/1040640165997191189/1174057055343100014/image.png?ex=656634d6&is=6553bfd6&hm=f9bbb950c8e9ec19d4ad5c39ef3e764b17f24a9fbda58adb34d25e3227e0cc3c& "Рисунок 3.")




<p  align="justify">  4. Для того, чтобы выполнить данную задачу, были зафиксированы все проделанные шаги в файле отчета и создана версия отчета в Git используя данные команды: git commit -m "Зафиксированы изменения в отчете" — чтобы зафиксировать изменения в репозитории с подходящим сообщением о коммите.
git log — для просмотра истории коммитов и подтверждения, что фиксация произведена. (см. Рис 4)
</p>

![Фиксирование версии отчёта](https://cdn.discordapp.com/attachments/1040640165997191189/1174062386915852308/image.png?ex=656639ce&is=6553c4ce&hm=225c60330955dba936a2c3cd4b29167a9b008a7b8f66f62d53a81076255b0ae8& "Рисунок 4.")




<p  align="justify">  5. Для того, чтобы выполнить данную задачу, была создана ветка отчета в Git с помощью команды: "git branch <имя ветки>", в которую были добавлены описания шагов, и создана версия отчета в этой новой ветке. С помощью команды: "git checkout <имя ветки>" осуществлялся переход между ветками (см. Рис 5)
</p>

![Создание новой ветки](https://cdn.discordapp.com/attachments/1040640165997191189/1174065442810253332/image.png?ex=65663ca6&is=6553c7a6&hm=87230eecbbf2defe78fddf4b4f45d0fe5e676c9c4f6a06076b0d5a92d4822547& "Рисунок 5.")




<p  align="justify">  6. Для того, чтобы выполнить данную задачу, было выполнено слияние последнего изменения отчета в основную ветку. С помощью данной команды "git checkout master" была сменена ветка, а потом с помощью команды: "git merge newbranch" было выполнено слияние с веткой отчёта (см. Рис 6)
</p>

![Слияние веток](https://cdn.discordapp.com/attachments/1040640165997191189/1174068625401987252/image.png?ex=65663f9d&is=6553ca9d&hm=bde6d8220e53d033ddede12f5baf13e382db1c386df35eff218044dfb44bd969& "Рисунок 6.")




<p  align="justify">  7. Для того чтобы выполнить данную задачу, я зарегистрировался на GitHub, создал новый проект под названием «homework IIT» и загрузил содержимое локального репозитория в проект на GitHub. С помощью команды: "git remote add origin <ссылка на репозиторий>" был связан локальный репозиторий с только что созданным на GitHub. Для того, чтобы загрузить содержимое локального репозитория была использована команда: "git push -u origin master"  (см. Рис 7)
</p>
